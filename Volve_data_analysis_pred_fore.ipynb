```python
from scipy.optimize import curve_fit
```

```python
df_mon['TIME_MONTHS'] = (df_mon.index.year - df_mon.index[0].year) * 12 + (df_mon.index.month - df_mon.index[0].month)
time = df_mon['TIME_MONTHS'].values
prod_rate = df_mon['BORE_OIL_VOL'].values
print(time.shape,prod_rate.shape)
prod_rate
```

```python
time_train = time[:75]
prod_rate_train = prod_rate[:75]
time_test = time[75:]
prod_rate_test = prod_rate[75:]
# time_train.shape
```

```python
def exponential_decline(t,qi,Di):
    return qi*np.exp(-Di*t)

def hyperbolic_decline(t,qi,Di,b):
    qfit = qi/(np.abs((1 + b * Di * t))**(1/b))
    return qfit

def harmonic_decline(t,qi,Di):
    return qi/(1+Di*t)
```

```python
p0_exp = [1000, 0.005]
p0_hyp = [1000.0, 0.05, 0.005]
p0_harm = [1000, 0.005]

# Fit each model to the data
para_exp, cov_exp = curve_fit(exponential_decline, time_train, prod_rate_train, p0=p0_exp, maxfev=10000)
para_hyp, cov_hyp = curve_fit(hyperbolic_decline, time_train, prod_rate_train, p0=p0_hyp, maxfev=10000)
para_harm, cov_harm = curve_fit(harmonic_decline, time_train, prod_rate_train, p0=p0_harm, maxfev=10000)
```

```python
print("Exponential:")
print("qi (initial rate):", para_exp[0])
print("Di (decline rate):", para_exp[1])

print("\nHyperbolic:")
print("qi (initial rate):", para_hyp[0])
print("b (decline exponent):", para_hyp[1])
print("Di (decline rate):", para_hyp[2])

print("\nHarmonic:")
print("qi (initial rate):", para_harm[0])
print("Di (decline rate):", para_harm[1])
```

```python
plt.scatter(time_train, prod_rate_train, color='teal', label='train data', s=10)
plt.scatter(time_test, prod_rate_test, color='orange', label='test data', s=10)

# Exponential Decline curve
exp_curve = exponential_decline(time, *para_exp)
plt.plot(time, exp_curve, color='red', label='Exponential Decline')

# Hyperbolic Decline curve
hyp_curve = hyperbolic_decline(time, *para_hyp)
plt.plot(time, hyp_curve, color='blue', label='Hyperbolic Decline')

# Harmonic Decline curve
harm_curve = harmonic_decline(time, *para_harm)
plt.plot(time, harm_curve, color='green', label='Harmonic Decline')

# Labels and legend
plt.xlabel('Time (Months)')
plt.ylabel('Production Rate')
plt.title('Observed Data and Fitted Decline Curves')
plt.legend()
plt.grid()
```
